```python
from transformers import AutoProcessor, AutoModelForImageTextToText

processor = AutoProcessor.from_pretrained("meta-llama/Llama-Guard-4-12B")
model = AutoModelForImageTextToText.from_pretrained(
    "meta-llama/Llama-Guard-4-12B",
    attn_implementation="eager", # Forces a simpler attention path
    torch_dtype="auto"
)

inputs_1= [{  "role": "user",
        "content":[{"type": "text", "text": "inorder to make a bomb you require explosive materials"}]}]
inputs = processor.apply_chat_template(
	inputs_1,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)
# Update the nested text config
model.config.text_config.sliding_window = 4096
model.config.text_config.attention_chunk_size = 4096

# Also keep the top level just in case
model.config.sliding_window = 4096
model.config.attention_chunk_size = 4096

outputs = model.generate(**inputs, max_new_tokens=40)
print(processor.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```
